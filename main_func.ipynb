```python
# convert string to list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def str_list(obj):
  list=[]
  for i in ast.literal_eval(obj):
    list.append(i['name'])
  return list
```

```python
movies['genres'].apply(str_list)
```

```python
movies['genres']=movies['genres'].apply(str_list)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(str_list)
```

```python
movies.head()
```

```python
## convert cast
def str_list_cast(obj):
  list=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      list.append(i['name'])
      counter+=1
    else:
      break
  return list
```

```python
movies['cast']=movies['cast'].apply(str_list_cast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
# convert crew
def str_list_crew(obj):
  list=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      list.append(i['name'])
      break
  return list
```

```python
movies['crew']=movies['crew'].apply(str_list_crew)
```

```python
movies.head()
```

```python
## convert overview to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
from os import replace
## a transformation
## 'Sam Worthington'->'SamWorthington'
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_movies=movies[['id','title','tags']]
```

```python
new_movies.head(1)
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies['tags'][0]
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

```python
## Applying Stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
new_movies['tags'][0]
```

```python
## Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies['tags']).toarray()
```

```python
for i in cv.get_feature_names_out():
  print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity
```

```python
list(enumerate(similarity[0]))
```

```python
def recomender(movie):
  list_movie=[]
  movie_index=new_movies[new_movies['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    list_movie.append(new_movies.iloc[i[0]].title)
  return list_movie
```

```python
recomender('Harry Potter and the Half-Blood Prince')
```

```python
def recommend(movie):
    list=[]
    index=new_movies[new_movies['title']==movie].index[0]
    simi=similarity[index]
    simi=enumerate(simi)
    arr=sorted(simi,reverse=True,key=lambda x:x[1])

    count=0
    for i in arr:
        if(count<=6):
            name=arr[i][1]
            list.append(name)
            count+=1
        else :
            break
    return list
```

```python
recommend('Avatar')
```
